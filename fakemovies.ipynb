```python
# combine fandango table with all sites table
# join the tables 
df = pd.merge(fandango, all_sites_scores, on="FILM", how="inner")
```

```python
df
```

```python
# Normalise the ratings so they are between 0 and 5 
# to convert ratings 100/20 = 5 10/2 = 5 
```

```python
df.describe().transpose()["max"]
```

```python
df["RT_Norm"] = np.round(df["RottenTomatoes"]/20,1) # rounding to the nearest decimal
df["RTU_Norm"] = np.round(df["RottenTomatoes_User"]/20,1)
```

```python
df["Meta_Norm"] = np.round(df["Metacritic"]/20,1)
df["Meta_Norm_User"] = np.round(df["Metacritic_User"]/2, 1)
```

```python
df["IMDB_Norm"] = np.round(df["IMDB"]/2,1)
```

```python
df
```

```python
df.columns
```

```python
# create column with normalised values 
norm_scores = df[['STARS', 'RATING','RT_Norm', 'RTU_Norm', 'Meta_Norm', 'Meta_Norm_User', 'IMDB_Norm']]# pass in the list of normalised scores
```

```python
norm_scores
```

```python
# Moment of truth 
# stars on the website are higher 
```

```python
# create a distribution plot 
plt.figure(figsize=(10,4), dpi=300)
sns.kdeplot(data=norm_scores, clip=[0,5], shade=True, palette="Set1")
```

```python
# Stars and Rating are much higher compared to other movie companies 
```

```python
# Uneven distribution of movies compared to the other movies [] ie Rotten Tomatoes 
```

```python
# worst movies across the sites using cluster map 
```

```python
sns.clustermap(norm_scores, cmap="magma",col_cluster=False) # Heatmap 
```
